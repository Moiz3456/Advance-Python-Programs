```python
#finding the max in the given list
a=[10,2,3,4,5,6,7,8]
def maxi(a):
    m=0
    for i in a:
        if i>m:
            m=i
    return m

#the above code doesn't work if all the values in the array are negative
```

```python
maxi(a)
```

```python
#finding the largest in the array
b=[-1,-2,-4]
def largest(b):
    my_max=b[0]
    for i in range(1,len(b)):
        if my_max<b[i]:
            my_max=b[i]
    return my_max
```

```python
largest(a)
```

```python
#finding the max in the given list
a=[10,2,3,4,5,6,7,8]
def maxi(a):
    m=0
    for i in a:
        if i>m:
            m=i
    return m

#the above code doesn't work if all the values in the array are negative
```

```python
maxi(a)
```

```python
#finding the largest in the array
b=[-1,-2,-4,10,15,82,21]
def largest(b):
    my_max=b[0]
    for i in range(1,len(b)):
        if my_max<b[i]:
            my_max=b[i]
    return my_max
```

```python
largest(b)
```

```python
#different approach in locating the largest value
def alternate(b):
    for i in b:
        i_is_largest=True
        for j in b:
            if i<j:
                i_is_largest=False
                break
        if i_is_largest:
            return i
    return None
```

```python
alternate(b)
```

```python
#finding the two largest values in a list
def largest_two(b):
    my_max,second=b[:2]
    if my_max<second:
        my_max,second=second,my_max
    for i in range(2,len(b)):
        if my_max<b[i]:
            my_max,second=b[i],my_max
        elif second<b[i]:
            second=b[i]
    return (my_max,second)
```

```python
largest_two(b)
```

```python
def sorting_two(b):
    return tuple(sorted(b,reverse=True)[:2])
```

```python
sorting_two(b)
```

```python
def double_max(b):
    my_max=max(b)
    copy=list(b)
    copy.remove(my_max)
    return (my_max,max(copy))
```

```python
double_max(b)
```
